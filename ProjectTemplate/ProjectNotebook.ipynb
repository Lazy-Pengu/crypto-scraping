# Project Description

This is webscraping program which allows the user to select from a list of cryptocurrency and obtain the current information about one of its attributes. This done through using the CoinMarketCap API which allows for requests to be made and information to be accessed. The user first is able to choose a cryptocurrency from the list displayed. Next, they can choose of of the attributed of the cryptocurrency that they want to get information from. Lastly, the user can choose to email the information to a specificed recipient or print the info out onto the screen. For each step, the user is allowed 3 attempts to enter in a vaild input.

```python
#Creates the UrlCreator object to access the CoinMarketCap API
API_access = UrlCreator()

#Gets the crypto name listings and prints it
crypt_name_list = API_access.get_name_listing()
print_listing(crypt_name_list)

#Checks to see if the user selection is within the list of cryptocurrencies
name_present = False
wrong_input_counter = 3
while (not name_present) and wrong_input_counter >= 0:
    crypt_name = input('\nSelect a Cryptocurrency: ').lower()

    #Searches for user seletion in cypto. list
    for name in crypt_name_list:
        
        #Breaks out of loop if name is present
        if name == crypt_name:
            name_present = True
            break
        else:
            name_present = False

    #Prints error message and decrease max tries by one
    if not name_present:
        if wrong_input_counter != 0 :
            print('Input is not present in Listing.\n' +
                  str(wrong_input_counter) + ' Tries left')
        wrong_input_counter -= 1
        
#Executes if user input valid crypto. name
if name_present:
    #Sets the final url to access specific info of the crypto.
    API_access.set_url(crypt_name)
    
    #Gets the attributes of the crypto and prints it
    crypt_keys = API_access.get_keys()
    print_keys(crypt_keys)
    
    
    key_present = False
    wrong_input_counter = 3
    #Checks to see if user input is valid
    while (not key_present) and wrong_input_counter >= 0:
        crypt_attribute = input('\nSelect a value: ').lower()

        #Breks out of loop if user selection is present in list
        for key in crypt_keys:
            if key == crypt_attribute:
                key_present = True
                break
            else:
                key_present = False
                
        #Prints error message and decrease max tries by one
        if not key_present:
            if wrong_input_counter != 0:
                print('Input is not present in Value.\n' +
                      str(wrong_input_counter) + ' Tries left')
            wrong_input_counter -= 1

    #Executes if user attribute selection is valid
    if key_present:
        #Obtains the requested info and store it
        answer = API_access.get_answer(crypt_attribute)
        crypt_info = '\n' + crypt_name + ' \n' +  '----------\n' + crypt_attribute + ': ' + answer
        
        end_program = False
        vaild_email_ans = False
        wrong_input_counter = 3
        #Checks to see if the user answer is valid
        while (not vaild_email_ans) and wrong_input_counter >= 0:
            send_info_ans = input('\nWould you like to email this information? ').lower()
            
            #Executes if user wants to send information
            if send_info_ans == 'yes':
                name, ident = send_email()

                #Creates EmailInfo object and sets up email server
                print("*** USING DEFAULT EMAIL ***")
                email_msg = EmailInfo()
                succeeded = email_msg.setup_server()
                
                valid_recepient_email = False
                second_wrong_counter = 3
                #Checks if ther user answer is valid
                while not valid_recepient_email and second_wrong_counter >= 0:
                    recepient_email = input('\nEnter email of recepient: ')

                    #Checks if the user inputted email contains required characters
                    if ('@' in recepient_email) and ('.com' in recepient_email):
                        valid_recepient_email = True
                    
                    #Prints error message and decrease max tries by one
                    if not valid_recepient_email:
                        if second_wrong_counter != 0:
                            print('Enter a valid Email address.\n' +
                                  str(second_wrong_counter) + ' Tries left')
                        else:
                            #Ends program is tries reaches 0
                            end_program = True
                        second_wrong_counter -= 1

                #Executes if the inputted recepient's email is vaild
                if valid_recepient_email:
                    #Sets the subject line of email to user input
                    subject = input('Enter subject of email: \n' +
                                    '*** DEFAULT WILL BE CRYPTOCURRENCY TYPE ***')
                    #Default to name of crypto. and attribute requested if no subject is entered
                    if subject == '':
                        subject = crypt_name + ' ' + crypt_attribute

                    #Sets up and sends the email message to the recepient
                    email_msg.setup_email(recepient_email, subject, crypt_info)
                    email_msg.send_email()

                    vaild_email_ans = True       
                    
                elif end_program:
                    #Breaks out of loop to end program
                    break;
                
            elif send_info_ans == 'no':
                #Prints info if user does not want to send 
                print( 'Your Info : \n' + '\033[1m' + crypt_info + '\033[0m')
                vaild_email_ans = True
                end_program = True

            #Prints error message and decrease max tries by one
            if not vaild_email_ans:
                if wrong_input_counter != 0:
                    print('Input is not present in Value.\n' +
                          str(wrong_input_counter) + ' Tries left')
                    print('Enter yes or no: ')
                else:
                    #Ends program is tries reaches 0
                    end_program = True
                wrong_input_counter -= 1   
                
        if end_program:
            #Prints Exit Message
            print('\n***Program Exited***')
    else:
        #Prints Exit Message
        print('\n***Program Exited***')
else:
    #Prints Exit Message
    print('\n***Program Exited***')
```
